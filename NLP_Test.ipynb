```python
import pandas as pd
df = pd.read_csv('éš¨æ„é³¥åœ°æ–¹101è§€æ™¯é¤å»³.csv')
df.head()
```

```python
import jieba
word = jieba.cut(df.iloc[0]['user_reviews'])
result = "  ".join(word)
print(result)
```

```python
words= []
for i,row in  df.iterrows():
    word = jieba.cut(row['user_reviews'])
    result = '  '.join(word)
    words.append(result)
print(words)
words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer() #å°†æ–‡æœ¬è½¬æ¢ä¸ºæ•°å€¼ï¼Œæ„æˆç‰¹å¾å‘é‡
X = vect.fit_transform(words)
X = X.toarray()
X
```

```python
words_bag = vect.vocabulary_
print(words_bag)
```
