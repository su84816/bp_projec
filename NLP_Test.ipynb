```python
import pandas as pd
df = pd.read_csv('隨意鳥地方101觀景餐廳.csv')
df.head()
```

```python
import jieba
word = jieba.cut(df.iloc[0]['user_reviews'])
result = "  ".join(word)
print(result)
```

```python
words= []
for i,row in  df.iterrows():
    word = jieba.cut(row['user_reviews'])
    result = '  '.join(word)
    words.append(result)
print(words)
words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer() #将文本转换为数值，构成特征向量
X = vect.fit_transform(words)
X = X.toarray()
X
```

```python
words_bag = vect.vocabulary_
print(words_bag)
```
