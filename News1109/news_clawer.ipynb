```python
import requests
from bs4 import BeautifulSoup
import json
import codecs
import time

url = 'https://news.pts.org.tw/category/1?page={page}' #目標網址>>公視/政治板 , page後面改就好
for i in range(1,36): #抓到35頁
  time.sleep(5) #每次爬蟲休息5秒
  print(f'-----------------------{i}----------------------------')
  r1 = requests.get(url.format(page=i)) #對目標做資料讀取
  # print(r1.text)
  s1 = BeautifulSoup(r1.text,"html.parser") #轉html格式

  for s2 in s1.find_all("h2"):

    title = s2.find("a").text
    # print(title) #抓標題,不能有/不然不能取檔名 >>> 後來發現他有兩種斜線= =
    title_del = "／/"
    for x in range(len(title_del)):
      new_title = title.replace(title_del[x],"")
    print(new_title)

    link = s2.find("a").attrs["href"]
    print(link) #抓連結 

    r2 = requests.get(s2.find("a").attrs["href"]) #把連結存到r2並用requests呼叫
    s3 = BeautifulSoup(r2.text,"html.parser") #s3存 r2轉html

    for s4 in s3.find_all("script",{"type":"application/ld+json"}): #從標籤就能看出來,裡面是json
      j1 = json.loads(s4.text) #這個j1並不夠乾淨,print target會有None產生,所以後面要用if篩選
      # print(j1)
      target = j1.get('articleBody')#目標："articleBody"
      if target != None:
        # print(target)
        final_target = target

      with codecs.open("/content/drive/MyDrive/2021-AI/2021-團體專題/AI02-第三組-輿情系統/Sam/News1109/(p"+str(i)+")"+str(new_title)+".csv",'w','utf-8') as f: #雲端存檔
        f.write("標題,連結,內文\n") #這裡在做欄位
        f.write(new_title)
        f.write(",")
        f.write(link)
        f.write(",")
        f.write(final_target) 
```
